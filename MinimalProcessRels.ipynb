```python
def indices(lst, element): #returns list containing indices of all instances of an element in an input list, passed input list and element to be searched for as parameters
    result = []
    offset = -1
    while True:
        try:
            offset = lst.index(element, offset+1) #determines how much of the input list the index method returns the first instance of the target element from, starts with whole list (offset+1 = 0), then from the element after the first instance and so on.
        except ValueError:
            return result #returns list of instances of target element when index method can no longer find an instance of the target element in the remainder of the list
        result.append(offset)
```

```python
def ProcessRels(path, title): #converts output rels file from MOD into a more readable form, takes filepath of raw rels file and file name for output file as parameters
    Rels_data = pd.read_csv(path, sep='\t')
    indexes = []
    rules = []
    for i in range(len(Rels_data['Index'])):
        if Rels_data['Index'][i] not in indexes:
            indexes.append(Rels_data['Index'][i])
            rules.append(Rels_data['Rule'][i]) #identifies new reactions (denoted by new indices) and the rule to which they correspond
        # if i%10000 == 0:
        #     print(i)
    print('Checkpoint')
    reactions = []
    reagents = []
    products = []
    for i in range(len(Rels_data['Index'])):
        reactions.append(Rels_data['Index'][i]) #creates a list of all reactions (repeated by how many molecules are involved in them)
    print('Checkpoint 2')
    print(len(indexes)) #prints the no. unique reactions
    for i in range(len(indexes)): #iterates over the unique reactions
        molecules = []
        coefficients = []
        dummy = indices(reactions, indexes[i]) #creates a list of all the instances of a particular reaction index in the total list of reactions
        for j in range(len(dummy)):
            molecules.append(Rels_data['Reagent'][dummy[j]])
            coefficients.append(Rels_data['Formed/Produced'][dummy[j]]) #for each instance of the reaction index, the associated molecule and coefficient (whether it is used(-1) or made(1)) is stored
        used = []
        made = []
        for k in range(len(coefficients)):
            if coefficients[k] == 1:
                made.append(molecules[k])
            else:
                used.append(molecules[k]) #each molecule associated with a particular reaction index is classified as used or made based on its coefficient
        reagents.append(used)
        products.append(made)
    data = {'Index': indexes, 'Reagents': reagents, 'Products': products, 'Rule': rules}
    df = pd.DataFrame(data)
    df.to_csv(f'/content/drive/MyDrive/BMSIS /MinimalDirectory/ProcessedData/RelsFiles/{title}ProcessedRels.tsv', header=None, index=None, sep='\t', mode='a')
    return(df)
```
