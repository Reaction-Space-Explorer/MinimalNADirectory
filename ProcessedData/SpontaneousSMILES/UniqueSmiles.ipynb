```python
import pandas as pd

df = pd.read_csv('AllSpontaneousSmiles.tsv', delimiter='\t')

grouped = df.groupby('Smiles')

result_df = pd.DataFrame(columns=['Smiles', 'Generation', 'Network'])

for smile, group in grouped:
    generations = group['Generation'].unique()
    networks = group['Network'].unique()
    
    generations_str = ', '.join(map(str, generations))
    networks_str = ', '.join(map(str, networks))
    
    result_df = result_df.append({'Smiles': smile, 'Generation': generations_str, 'Network': networks_str}, ignore_index=True)

result_df = result_df.sort_values(by='Smiles')

result_df.to_csv('UniqueSmiles.tsv', sep='\t', index=False)
```
