```python
def spontaneous_filter(Network): #function to remove non-spontaneous reactions from an input rels file, passed name of network
    df = pd.read_csv(f'/content/drive/MyDrive/BMSIS /MinimalDirectory/ProcessedData/RelsWithThermoFiles/{Network}G3RelsWithThermo.tsv', sep='\t')
    df = df.dropna() #removes any reactions with no evaluated energy because thermodynamic data cannot be generated using equilibrator_assets
    df = df.drop(list(df[df['Energy Change'] >= 0].index)) #removes all reactions with a positive change in reaction Gibbs energy (i.e non-spontaneous reactions)
    df.to_csv(f'/content/drive/MyDrive/BMSIS /MinimalDirectory/ProcessedData/SpontaneousRelsWithThermoFiles/Spontaneous{Network}G3RelsWithThermo.tsv', index=None, sep='\t')
    return(df)
```
